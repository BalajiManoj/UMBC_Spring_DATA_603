1. Write Python code and use MapReduct to count occurrences of each word in the first text file (file.txt). How many times each word is repeated?

```python
import re
import pandas as pdS
import numpy as np
from itertools import groupby
import enchant
import string

def maping_function(key, val): #word counting 
    word_count = [x[1] for x in val]
    return [key,sum(word_count)]

def data_map_clean(var): # mapping and data cleaning
    data_cleanup = ''.join([i for i in var.lower()])
    res = re.sub(r'[^\w\s]', '', data_cleanup)
    return [data_cleanup,1]

def remove_punctuation(lst):
    table = str.maketrans("", "", string.punctuation) 
    temp_list = []
    for s in lst:
        char=s.translate(table)
        temp_list.append(char.replace(',', '').replace('"', '').replace('.', '').replace('”', '').replace('“', '').replace('’', '').replace(' ', ''))
    return temp_list
```

```python
# seperating the words in the text file
file = open('Harry_Potter_Series_Book_1.txt', encoding="utf8")
df = [word for line in file for word in line.split()]
print("Extracted Data:")
print (df)

# cleating data with revoing punctuation
df = remove_punctuation(df)

# using the map function for the words
map_data = map(data_map_clean, df)
#print(map_data)

# sorting and counting
result = [];
for i, j in groupby(sorted(map_data), key = lambda x: x[0]):
    result.append(maping_function(i, list(j)))
    
print('\n')
print("Mapped Data:")  
print(result)    
```
